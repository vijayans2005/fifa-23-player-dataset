```python
# Creating a mock FIFA 23 player dataset
data = {
    'Player Name': ['Lionel Messi', 'Cristiano Ronaldo', 'Neymar', 'Kylian Mbappe', 'Mohamed Salah',
                    'Kevin De Bruyne', 'Robert Lewandowski', 'Luka Modric', 'Karim Benzema', 'Virgil van Dijk'],
    'Age': [36, 39, 32, 25, 32, 33, 35, 38, 36, 33],
    'Overall': [91, 90, 89, 91, 89, 91, 91, 87, 89, 90],
    'Position': ['RW', 'CF', 'LW', 'CF', 'RW', 'CM', 'CF', 'CM', 'CF', 'CB'],  # RW: Right Wing, CF: Center Forward, etc.
    'Club': ['PSG', 'Al Nassr', 'PSG', 'PSG', 'Liverpool', 'Manchester City', 'Barcelona', 'Real Madrid', 'Al Ittihad', 'Liverpool'],
    'Country': ['Argentina', 'Portugal', 'Brazil', 'France', 'Egypt', 'Belgium', 'Poland', 'Croatia', 'Saudi Arabia', 'Egypt'],
    'Market Value (in million €)': [50, 20, 80, 180, 100, 120, 50, 40, 10, 50],
    'Goals': [799, 840, 400, 300, 280, 110, 600, 100, 500, 200]
}

# Create DataFrame
fifa_df = pd.DataFrame(data)

# Display the DataFrame
fifa_df
```

```python
# Visualize Age Distribution using a Histogram
plt.figure(figsize=(10, 5))
sns.histplot(fifa_df['Age'], kde=True, color='blue', bins=10)  # Increase bins for better granularity
plt.title('Age Distribution of FIFA 23 Players')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualize the Market Value Distribution using a Boxplot
plt.figure(figsize=(10, 5))
sns.boxplot(x=fifa_df['Market Value (in million €)'], color='green')
plt.title('Market Value Distribution of FIFA 23 Players')
plt.xlabel('Market Value (in million €)')
plt.show()
```

```python
# Pie Chart for Player Positions
position_counts = fifa_df['Position'].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(position_counts, labels=position_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightgreen', 'lightskyblue', 'lightyellow', 'lightpink'])
plt.title('Distribution of Player Positions in FIFA 23')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()
```

```python
# Scatter Plot for Goals vs Market Value
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Goals', y='Market Value (in million €)', data=fifa_df, hue='Country', palette='Set2', s=100)
plt.title('Goals vs Market Value (Colored by Country)')
plt.xlabel('Goals Scored')
plt.ylabel('Market Value (in million €)')
plt.legend(title='Country')
plt.show()
```

```python
# Correlation Heatmap for numeric columns
plt.figure(figsize=(10, 6))
sns.heatmap(fifa_df[['Age', 'Overall', 'Market Value (in million €)', 'Goals']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of FIFA 23 Players')
plt.show()
```
